```python
import numpy as np
import pandas as pd
from openai import OpenAI
import time

# Initialize the client
client = OpenAI(api_key="sk-093306576cf54aa78d45a5515041b388", base_url="https://api.deepseek.com")
```

# 9.i1 - 9.9

```python
# Function to get calculation result from API
def get_calculation_result(expression):
    try:
        response = client.chat.completions.create(
            model="deepseek-chat",
            messages=[
                {"role": "system", "content": "You are a helpful assistant"},
                {"role": "user", "content": f"{expression} = ? Only answer the result, without any explanatory text."},
            ],
            stream=False
        )
        return response.choices[0].message.content.strip()
    except Exception as e:
        return f"Error: {e}"

# Create the test matrix
expressions = []
results_matrix = []

print("Starting calculations...")

# Loop through i from 0 to 9
for i in range(10):
    expression = f"9.{i}1 - 9.9"
    expressions.append(expression)
    
    print(f"Testing {expression}...")
    
    # Test 5 times for each expression
    trial_results = []
    for trial in range(5):
        result = get_calculation_result(expression)
        trial_results.append(result)
        print(f"  Trial {trial + 1}: {result}")
        time.sleep(0.5)  # Small delay to avoid rate limiting
    
    results_matrix.append(trial_results)

# Convert to numpy array
results_array = np.array(results_matrix)

# Create DataFrame for better display
column_names = [f"Trial {i+1}" for i in range(5)]
df = pd.DataFrame(results_array, index=expressions, columns=column_names)

print("\nFinal Results Matrix:")
print(df)

# Save as numpy array if needed
np.save('calculation_results.npy', results_array)
print("\nResults saved to 'calculation_results.npy'")
```

# 9.11 - 9.i

```python
# Function to get calculation result from API
def get_calculation_result(expression, trial_num):
    try:
        system_message = "You are a helpful assistant"
        user_message = f"{expression} = ? Only answer the result, without any explanatory text."
        
        # Print the prompt for trial 1
        if trial_num == 1:
            print(f"    PROMPT for Trial 1:")
            print(f"    System: {system_message}")
            print(f"    User: {user_message}")
            print()
        
        response = client.chat.completions.create(
            model="deepseek-chat",
            messages=[
                {"role": "system", "content": system_message},
                {"role": "user", "content": user_message},
            ],
            stream=False
        )
        return response.choices[0].message.content.strip()
    except Exception as e:
        return f"Error: {e}"

# Create the test matrix
expressions = []
results_matrix = []

print("Starting calculations...")

# Loop through i from 0 to 9
for i in range(10):
    expression = f"9.11 - 9.{i}"
    expressions.append(expression)
    
    print(f"Testing {expression}...")
    
    # Test 5 times for each expression
    trial_results = []
    for trial in range(5):
        result = get_calculation_result(expression, trial + 1)
        trial_results.append(result)
        print(f"  Trial {trial + 1}: {result}")
        time.sleep(0.5)  # Small delay to avoid rate limiting
    
    results_matrix.append(trial_results)

# Convert to numpy array
results_array = np.array(results_matrix)

# Create DataFrame for better display
column_names = [f"Trial {i+1}" for i in range(5)]
df = pd.DataFrame(results_array, index=expressions, columns=column_names)

print("\nFinal Results Matrix:")
print(df)

# save df as csv
df.to_csv('calculation_results.csv', index_label='Expression')

# Save as numpy array if needed
np.save('calculation_results.npy', results_array)
print("\nResults saved to 'calculation_results.npy'")
```
